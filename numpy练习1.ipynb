```python
import numpy as np
```

```python
a = np.random.randint(0,100,12)
print(a)
b = np.reshape(a,(3,4))
print(b)
```

```python
print('每一列的平均值为:',np.mean(b,axis=0))
for i in range(b.shape[1]):
    print(f"第 {i + 1} 列的平均值为：{np.mean(b[:, i]):.2f}")
```

```python
print('每一行的平均值为:',np.mean(b,axis=1))
for i in range(b.shape[0]):
    print(f'第{i + 1}行的平均值为:{np.mean(b[i,:]):.2f}')
```

```python
a = np.random.randint(0,20,(5,5))
print(a)
print(a[a>10])
print(np.where(a>10,0,a))
```

```python
money = np.array([120,135,110,125,135,140])
print(money)
print('最高的月份是:',np.argmax(money)+1)
print('最低的月份是:',np.argmin(money)+1)
```

```python
a = np.array([1,2,3])
print(a)
b = np.array([4,5,6])
print(b)
c = np.concatenate([a,b]) #矩阵拼接
print(c)
```

```python
a = np.array([1,3,2,1,3,3,5,2,1])
unique_values,counts=np.unique(a, return_counts=True)
print('唯一值分别是:',unique_values)
print('重复次数分别是:',counts)
for value,count in zip(unique_values,counts):
    print(f'{value}出现了{count}次')
```
