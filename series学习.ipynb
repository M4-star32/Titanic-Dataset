```python
import pandas as pd
a = pd.Series([1,2,3,4,5])
print(a)
b = pd.Series([2,5,3,9,10],index=['a','b','c','d','e'])
print(b)
```

```python
a = pd.Series([1,9,54,23,7,20],index=['a','b','c','d','e','f'])
print(a)
a1 = pd.Series(a,index=['a','c'])
print(a1)
```

```python
print(a.index)
print(a.values)
print(a.shape,a.ndim,a.size)
a.name='score'
print(a.name)
print(a.loc['a':'b']) # loc索引支持切片
print(a.at['d']) # at索引不支持切片
```

```python
print(a)
print(a[a>50])
print(a.head(3))
print(a.tail(1))  #输出表头三个数字，表尾最后一个数字
```

```python
c = pd.Series([1,6,None,None,23,18],index=['a','b','c','d','e','f'],name='core')
print(c.head(2))
print(c.describe())
print(c.keys())
print(c.index) #获取索引，keys是方法需要加括号，index是属性不用加括号
print(c.isna()) #判断是否是缺失值
```

```python
print(c)
print(c.isin([1,6]))
print(c.var())
print(c.min())
print(c.median())
```

```python
c1 = c.sort_values() #排序
print(c1)
print(c.value_counts())
```

```python
c = pd.Series([1,1,1,2,5,3])
print(c.drop_duplicates())
print(c.unique())
```
