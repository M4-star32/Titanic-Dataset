```python
import numpy as np
```

```python
print(np.sqrt(4))
a = np.array([1,4,9])
print(np.sqrt(a))
print(np.exp(1))
print(np.log(2.718))
```

```python
#计算三角函数
print(np.sin(np.pi/2))
print(np.cos(np.pi))

```

```python
#绝对值、次幂、五舍六入
a = np.array([[-1,-3,-6],[2,-8,-3]])
print(np.abs(a))
print(np.power(a,3))
b = np.array([3.4,4.5,6.7])
print(np.round(b))  #五舍六入
print(np.ceil(b))  #向上和向下取整
print(np.floor(b))
```

```python
a = np.array([1,2,4,5])
print(np.sum(a))
b = np.random.randint(1,100,5)
print(b)
print(np.average(b))
print(np.median(b))
print(np.percentile(a,25))  # 有25%的元素比1.75小
```

```python
print(np.var(b))
print(np.std([1,2,3]))
```

```python
a = np.random.randint(1,100,8)
print(a)
print(np.max(a),np.argmax(a))
print(np.min(a),np.argmin(a))
```
