```python
import numpy as np
```

```python
a = np.random.randint(1, 100,10
print(a)
print(np.greater(a,10))  #是否大于 小于 等于
print(np.less(a,10))
print(np.equal(a,10))
```

```python
print(np.any([0,0,0,0]))
print(np.any([0,0,0,1])) #检查是否有一部分元素为真
print(np.all([0,0,0,0,0]))
print(np.all([1,1,1,1])) #检查是否全部元素为真
```

```python
print(a)
print(np.where(a>10,a,0)) #从a中找大于10的元素，如果不满足条件则为0  条件 符合条件的 不符合条件的
print(np.where(a>50,1,0))
```

```python
print(a)
print(np.where(a<60,'不及格',np.where(a<80,'良好','优秀')))
print(np.select([a<60,(a>=60) & (a<=80),a>=80],['不及格','良好','优秀'],default = '未知'))
```

```python
b = np.random.randint(1,100,10)
print(b)
print(np.sort(b),np.argsort(b),sep = '\n')
print(np.unique(b)) # 去重，会自动排序
```

```python
print(a)
print(b)
print(np.concatenate((a,b)))
print(np.empty_like(a))
```
