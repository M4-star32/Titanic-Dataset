# 一、ndarray
## ndarray的特性
### 1. 多维性

```python
import numpy as np
```

```python
a = np.array([1,2,3])
print(a)
print('数组的形状:',a.shape)
print('数组的维度:',a.ndim)
print('数组的个数:',a.size)
print('数据类型:',a.dtype)
```

```python
a = np.array([ [1,2,3],[2,3,4] ])
print(a)
print('数组的形状:',a.shape)
print('数组的维度:',a.ndim)
print('数组的个数:',a.size)
```

```python
list = [2,3],[3,4]
a = np.array(list,dtype=int)
print(a)
print('数组形状:',a.shape)
print('数组维度:',a.ndim)
```

```python
b = np.copy(a)
print(b)
b[1,0] = 6
print(b)
```

```python
c = np.zeros(((3,2)),dtype=int)
print(c)
d = np.zeros((5,1),dtype = int)
print(d)
```

```python
a = np.full((3,4),5)
print(a)
```

```python
a = np.arange(1,10,2)  #左闭右开，开始，结束，步长，不包含结束
print(a)
b = np.linspace(1,10,5)  # linespace参数：开始，结束，步长会自动算出间隔
print(b)
```

```python
#等比数列
a = np.logspace(0,8,3,base=4,dtype=int)
print(a)

b = np.logspace(0,8,3,dtype=int)
print(b)
```

```python
a = np.eye(3)
b = np.eye(3,dtype = int)
print(a)
print(b)
```

```python
a = np.diag([3,4,'b',3])
print(a)
```

```python
# 随机生成浮点数矩阵
a = np.random.rand(2,5)  # 生成0-1区间内的2x5矩阵
print(a)
b = np.random.uniform(2,7,(3,4)) #生成2-7之间的3x4矩阵
print(b)
```

```python
a = np.random.randint(3,5,(3,8))
print(a)
```

```python
a = np.array([1,2,0],dtype=bool)
print(a)
b = np.array([-128,127],dtype=np.int8)
print(b)
```
