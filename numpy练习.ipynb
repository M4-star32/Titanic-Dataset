```python
import numpy as np
```

```python
a = np.array([28,30,29,31,32,30,29])
print(a)
print('平均气温是:',np.round(np.mean(a))) #五舍六入
print('平均气温是:%.3f'%np.mean(a))   # 保留三位小数
print('最高温度是:',np.max(a))
print('最低温度是:',np.min(a))
print('气温超过30的天数:',len(a[a>30]))
print('气温超过30的天数:',np.sum(np.where(a>30,1,0)))
```

```python
a = np.random.randint(0,10,(3,4))
print(a)
for i in range(0,4):
    print(f'第{i+1}列的最大值为:',np.max(a[:,i]))
    i +=1
for i in range(0,3):
    print(f'第{i+1}行的最大值为',np.max(a[i,:]))
    i +=1
```

```python
a = np.random.randint(0,10,(3,4))
print(a)
print('每一列的最大值为',np.max(a,axis=0))
print('每一行的最大值为',np.max(a,axis=1))
print(np.where(a%2==1,-1,a))
```
